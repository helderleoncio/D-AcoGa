```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Importação das bibliotecas Graficas
import seaborn as sns
import scikitplot as skplt
import matplotlib.pyplot as plt

# Exibe gráficos na página (inline)
%matplotlib inline
```

### Dataset  100%

```python
df100 = pd.read_csv('../../../../datasets/dacoga/nbaiot/full/dsetN-Baiot_v2_100.csv')
df100.head()
```

```python
# Separando X e y, assumindo que a última coluna é y
X100 = df100.iloc[:, :-1]  # Todas as colunas exceto a última
y100 = df100.iloc[:, -1]   # Apenas a última coluna
```

```python
# Dividindo os dados em conjuntos de treino e teste (70% treino, 30% teste)
X100_train, X100_test, y100_train, y100_test = train_test_split(X100, y100, test_size=0.3, random_state=42)
```

```python
# Criando DataFrames para treino e teste incluindo tanto X quanto y
train_df = pd.concat([X100_train, y100_train], axis=1)
test_df = pd.concat([X100_test, y100_test], axis=1)
```

```python
diretorioTrain = '../../../../datasets/dacoga/nbaiot/train/'
arquivo_saida_train = 'dsetN-Baiot_v2_100_train.csv'
caminho_completo_saida = os.path.join(diretorioTrain, arquivo_saida_train)
train_df.to_csv(caminho_completo_saida, index=False)

diretorioTest = '../../../../datasets/dacoga/nbaiot/test/'
arquivo_saida_test = 'dsetN-Baiot_v2_100_test.csv'
caminho_completo_saida = os.path.join(diretorioTest, arquivo_saida_test)
test_df.to_csv(caminho_completo_saida, index=False)
```

```python
df100train = pd.read_csv('../../../../datasets/dacoga/nbaiot/train/dsetN-Baiot_v2_100_train.csv')
df100train.head()
```

```python
# Verificar o balanceamento das classes
#print(df.label.value_counts())
print('Label')
print('Ataques' , df100train.label.value_counts().index.tolist()[0], df100train.label.value_counts()[0])
print('Normal' , df100train.label.value_counts().index.tolist()[1], df100train.label.value_counts()[1])

print("\nAtaques representam {:.4f}% do dataset.\n".
      format((df100train[df100train.label == 0].shape[0] / df100train.shape[0]) * 100))

# plotar gráfico de barras para as Classes
sns.countplot(x=df100train['label'])
```

```python
print('Return a tuple representing the dimensionality of the DataFrame')
print('Colunas', df100train.shape[1])
print('Linhas', df100train.shape[0])
```
