## Funções do Sistema

```python
def get_attacks(dir, dev_name, attack, label):
    
    # Lista para armazenar os DataFrames processados
    dfs_processados = []

    # Percorre todos os arquivos no diretório
    for file in os.listdir(dir):
        # Verifica se o arquivo tem a extensão .csv
        if file.endswith('.csv'):
            # Caminho completo do arquivo
            full_path = os.path.join(dir, file)
            # Lê o arquivo CSV e armazena em um DataFrame
            df = pd.read_csv(full_path)
            
            # Adiciona colunas fixas
            df['Device_Name'] = dev_name
            df['Attack'] = attack
            
            # Adiciona uma nova coluna com o nome do arquivo (sem a extensão .csv)
            df['Attack_subType'] = os.path.splitext(file)[0]
            
            # Adiciona a coluna 'label'
            df['label'] = label
            
            # Adiciona o DataFrame processado à lista
            dfs_processados.append(df)

    return dfs_processados
```

```python
def get_normal(dir, dev_name, attack, label):

    # Lista para armazenar os DataFrames processados
    df_normal = []

    # Construindo o caminho completo do arquivo de forma segura
    file_path = os.path.join(dir, 'benign_traffic.csv')

    # Lendo o arquivo CSV
    df = pd.read_csv(file_path)
    
    # Adicionando as colunas com os valores dos parâmetros
    df['Device_Name'] = dev_name  # Usando o parâmetro dev_name
    df['Attack'] = attack  # Usando o parâmetro attack
    df['Attack_subType'] = 'Normal'  # Subcategoria, fixa como 'Normal'
    df['label'] = label  # Usando o parâmetro label

    # Adicionando o DataFrame configurado à lista
    df_normal.append(df)

    return df_normal
```

## Lendo dataset de trafego normal

```python
source = '../datasets/detectBbaiot/Philips_B120N10_Baby_Monitor/'
device = 'Philips_B120N10_Baby_Monitor'
attack = 'Normal'
label = 'Normal'

normal_temp = get_normal(source, device, attack, label)
normal = pd.concat(normal_temp, ignore_index=True)
normal.head()
```

## Tratando dataset ataques - Gafgyt

```python
source = '../datasets/detectBbaiot/Philips_B120N10_Baby_Monitor/gafgyt_attacks/'
device = 'Philips_B120N10_Baby_Monitor'
attack = 'gafgyt'
label = 'Anomaly'

gafgyt_temp = get_attacks(source, device, attack, label)
gafgyt = pd.concat(gafgyt_temp, ignore_index=True)
gafgyt.head()
```

### Concatenando os dataframes

```python
df_Temp = pd.concat([normal, gafgyt], ignore_index=True)

#Embaralhando as linhas do DataFrame com base no índice aleatório
df= df_Temp.sample(frac=1).reset_index(drop=True)
df.head()
```

```python
df.to_csv('../datasets/dacoga/dsetBM.csv', index=False)
```
